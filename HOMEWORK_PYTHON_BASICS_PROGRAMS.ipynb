## Program 23: Largest and Smallest in a List

**Problem**: Read `N` and a list of `N` integers. Print the smallest and largest.

**Input**:
- Integer `N`
- Next line: `N` integers

**Output**:
- Two integers: `min max`

**Sample Input**:
```
5
5 1 9 2 7
```
**Sample Output**:
```
1 9
```


````python
n = int(input("Enter the number of elements: "))

numbers = []

for i in range(n):
    numbers.append(int(input("Enter number: ")))

smallest = numbers[0]
largest = numbers[0]

for num in numbers:
    if num < smallest:
        smallest = num
    if num > largest:
        largest = num

print("Smallest =", smallest)
print("Largest =", largest)
````

## Program 24: Count Occurrences in a List

**Problem**: Read `N`, then `N` integers, then a target `x`. Print how many times `x` appears.

**Input**:
- Integer `N`
- Next line: `N` integers
- Next line: integer `x`

**Output**:
- Count of `x`

**Sample Input**:
```
6
1 2 2 3 2 4
2
```
**Sample Output**:
```
3
```


````python
n = int(input("Enter N: "))

numbers = []

for i in range(n):
    numbers.append(int(input("Enter number: ")))

x = int(input("Enter target x: "))

count = 0

for num in numbers:
    if num == x:
        count += 1

print("Total Count of the number =", count)
````
